```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

import ttsxai
```

```python
save_dir = os.path.join(
    ttsxai.PACKAGE_DIR, 
    "data_df/LJSpeech/tacotron2_waveglow"
)
```

```python
df = pd.read_pickle(os.path.join(save_dir, 'train_activation_df.pkl'))
```

```python
# Analyzing and visualizing distributions for all phones together
features = ['duration', 'pitch', 'energy']

for feature in features:
    # Collect data for all phones
    all_data = []
    labels = []
    unique_phones = np.unique(df['phone'].to_numpy())
    for phone in unique_phones:
        filtered_df = df[df['phone'] == phone]
        data = filtered_df[feature].to_numpy()
        all_data.extend(data)
        labels.extend([phone] * len(data))

    # Boxplot
    plt.figure(figsize=(20, 10))
    sns.boxplot(x=labels, y=all_data)
    plt.title(f"{feature.capitalize()} distribution for all phones")
    plt.xticks(rotation=90)  # Rotate x labels for better readability
    plt.ylabel(feature)
    plt.xlabel("Phones")
    plt.tight_layout()
    plt.show()
```
