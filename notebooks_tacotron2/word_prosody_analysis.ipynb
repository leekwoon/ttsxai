```python
with open('../data_analysis/analyze_word_prosody/word_data.pkl', "rb") as file:
    data_dict = cloudpickle.load(file)
```

```python
count_dict = {}
for word in data_dict.keys():
    count_dict[word] = len(data_dict[word]['duration'])
```

```python
# value에 따라 내림차순으로 정렬
sorted_data = sorted(count_dict.items(), key=lambda x: x[1], reverse=True)
# 정렬된 key 리스트, 길이는 최소 4인것만
sorted_keys = [item[0] for item in sorted_data if len(item[0]) >= 4]
# top-20
sorted_keys = sorted_keys[:20]
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

# Analyzing and visualizing distributions for all phones together
# features = ['duration']
features = ['duration', 'pitch', 'energy']

for feature in features:
    # Collect data for all phones
    all_data = []
    labels = []
    for word, data in data_dict.items():
        if word in sorted_keys:
        # phone = _id_to_symbol[int(token)].replace('@', '').lower()
            all_data.extend(data[feature])
            labels.extend([word] * len(data[feature]))

    # Boxplot
    plt.figure(figsize=(20, 10))
    sns.boxplot(x=labels, y=all_data)
    plt.title(f"{feature.capitalize()} distribution for all phones")
    plt.xticks(rotation=90, fontsize=15)  # Rotate x labels for better readability
    plt.ylabel(feature)
    plt.xlabel("Phones")
    plt.tight_layout()
    plt.show()
```

`have` duration이 가장 긴거랑 작은거 들어보기

```python
hop_length = 256
sampling_rate = 22050
```

```python
def compute_cumulative_sums(duration):
    out = [0]
    for d in duration:
        out.append(d + out[-1])
    return out
```

```python
idx = np.argmax(data_dict['have']['duration'])

pkl_file = data_dict['have']['pkl_file'][idx]
with open(pkl_file, "rb") as file:
    data = cloudpickle.load(file)

text = data['text']
wave = data['wave']
duration = data['duration']
word_duration = data_dict['have']['duration'][idx]
word_index_boundary = data_dict['have']['word_index_boundary'][idx]
print('pkl_file=', pkl_file)
print('word_duration=', word_duration)

wave_start = int(hop_length * compute_cumulative_sums(duration)[word_index_boundary[0]])
wave_end = int(hop_length * compute_cumulative_sums(duration)[word_index_boundary[1]])
wave_cut = wave[wave_start: wave_end]

ipd.Audio(wave_cut, rate=sampling_rate)
```
