```python
import cloudpickle
import numpy as np
import IPython.display as ipd
from matplotlib import pyplot as plt
```

```python
with open('../data_analysis/analyze_phone_prosody/phone_data.pkl', "rb") as file:
    data_dict = cloudpickle.load(file)
```

```python
# Analysis and Visualization
for phone, data in data_dict.items():
    print(f"Phone: {phone}")
    print(f"Duration - Mean: {np.mean(data['duration'])}, Std: {np.std(data['duration'])}")
    print(f"Pitch - Mean: {np.mean(data['pitch'])}, Std: {np.std(data['pitch'])}")
    print(f"Energy - Mean: {np.mean(data['energy'])}, Std: {np.std(data['energy'])}")
    print("------")

    # Displaying histograms for the distributions
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    ax[0].hist(data['duration'], bins=30, color='blue', alpha=0.7)
    ax[0].set_title(f"Duration distribution for Phone {phone}")
    ax[1].hist(data['pitch'], bins=30, color='green', alpha=0.7)
    ax[1].set_title(f"Pitch distribution for Phone {phone}")
    ax[2].hist(data['energy'], bins=30, color='red', alpha=0.7)
    ax[2].set_title(f"Energy distribution for Phone {phone}")

    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns

# Analyzing and visualizing distributions for all phones together
features = ['duration', 'pitch', 'energy']

for feature in features:
    # Collect data for all phones
    all_data = []
    labels = []
    for phone, data in data_dict.items():
        all_data.extend(data[feature])
        labels.extend([phone] * len(data[feature]))
    
    # Boxplot
    plt.figure(figsize=(20, 10))
    sns.boxplot(x=labels, y=all_data)
    plt.title(f"{feature.capitalize()} distribution for all phones")
    plt.xticks(rotation=90)  # Rotate x labels for better readability
    plt.ylabel(feature)
    plt.xlabel("Phones")
    plt.tight_layout()
    plt.show()
```
